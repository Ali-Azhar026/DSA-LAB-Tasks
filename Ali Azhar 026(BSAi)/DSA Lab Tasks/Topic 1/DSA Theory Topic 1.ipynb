```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    
    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])
    
    return merge(left_half, right_half)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    result.extend(left[i:])
    result.extend(right[j:])
    
    return result

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    
    pivot = arr[0]
    less = [x for x in arr[1:] if x <= pivot]
    greater = [x for x in arr[1:] if x > pivot]
    
    return quick_sort(less) + [pivot] + quick_sort(greater)


arr = [64, 34, 25, 12, 22, 11, 90]

print("Bubble Sort:", bubble_sort(arr.copy()))
print("Merge Sort:", merge_sort(arr.copy()))
print("Quick Sort:", quick_sort(arr.copy()))
```

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def fibonacci_recursive_memo(n):
    if n <= 1:
        return n
    return fibonacci_recursive_memo(n - 1) + fibonacci_recursive_memo(n - 2)

def fibonacci_recursive_manual(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        memo[n] = n
    else:
        memo[n] = fibonacci_recursive_manual(n - 1, memo) + fibonacci_recursive_manual(n - 2, memo)
    return memo[n]

print(fibonacci_recursive_memo(40))         
print(fibonacci_recursive_manual(40))       
```

```python
import time
import matplotlib.pyplot as plt
from functools import lru_cache

# Naive Recursive Fibonacci (O(2^n))
def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

@lru_cache(maxsize=None)
def fibonacci_recursive_memo(n):
    if n <= 1:
        return n
    return fibonacci_recursive_memo(n - 1) + fibonacci_recursive_memo(n - 2)

def fibonacci_iterative(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

test_values = [10, 20, 25, 30, 35, 40]
timings = {"Recursive": [], "Memoized Recursive": [], "Iterative": []}

for n in test_values:
    start = time.time()
    fibonacci_recursive(n)
    timings["Recursive"].append(time.time() - start)

    start = time.time()
    fibonacci_recursive_memo(n)
    timings["Memoized Recursive"].append(time.time() - start)

    start = time.time()
    fibonacci_iterative(n)
    timings["Iterative"].append(time.time() - start)

plt.figure(figsize=(10, 6))
plt.plot(test_values, timings["Recursive"], label="Naive Recursive (O(2^n))", marker='o')
plt.plot(test_values, timings["Memoized Recursive"], label="Memoized Recursive (O(n))", marker='o')
plt.plot(test_values, timings["Iterative"], label="Iterative (O(n))", marker='o')

plt.title("Fibonacci Performance Comparison")
plt.xlabel("n (Fibonacci Index)")
plt.ylabel("Execution Time (seconds)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

for i, n in enumerate(test_values):
    print(f"n = {n}")
    print(f"  Naive Recursive: {timings['Recursive'][i]:.6f} seconds")
    print(f"  Memoized Recursive: {timings['Memoized Recursive'][i]:.6f} seconds")
    print(f"  Iterative: {timings['Iterative'][i]:.6f} seconds")
```

```python
import time
import math

# O(1) - Constant time complexity
def constant_time_example(n):
    return 42  # Just a constant value (no operation depending on n)

# O(log n) - Logarithmic time complexity
def log_time_example(n):
    return math.log(n)  # Logarithmic operation (returns the logarithm of n)

# O(n) - Linear time complexity
def linear_time_example(n):
    total = 0
    for i in range(n):
        total += i
    return total  # Sum of all numbers from 0 to n-1

# O(n log n) - Linearithmic time complexity
def n_log_n_time_example(n):
    return sorted(range(n))  # Sorting operation (O(n log n))

# O(n^2) - Quadratic time complexity
def quadratic_time_example(n):
    count = 0
    for i in range(n):
        for j in range(n):
            count += 1
    return count  # Counting pairs (nested loops)

# O(2^n) - Exponential time complexity (only works for small n)
def exponential_time_example(n):
    if n <= 1:
        return n
    return exponential_time_example(n - 1) + exponential_time_example(n - 2)

# Testing function to measure execution time
def measure_execution_times():
    input_sizes = list(range(1, 1001))  # Input sizes from 1 to 1000
    results = {
        "Constant": [],
        "Logarithmic": [],
        "Linear": [],
        "N Log N": [],
        "Quadratic": [],
        "Exponential": []
    }

    for n in input_sizes:
        # Time the O(1) function (constant time)
        start_time = time.time()
        constant_time_example(n)
        results["Constant"].append(time.time() - start_time)

        # Time the O(log n) function (logarithmic time)
        start_time = time.time()
        log_time_example(n)
        results["Logarithmic"].append(time.time() - start_time)

        # Time the O(n) function (linear time)
        start_time = time.time()
        linear_time_example(n)
        results["Linear"].append(time.time() - start_time)

        # Time the O(n log n) function (n log n time)
        start_time = time.time()
        n_log_n_time_example(n)
        results["N Log N"].append(time.time() - start_time)

        # Time the O(n^2) function (quadratic time)
        start_time = time.time()
        quadratic_time_example(n)
        results["Quadratic"].append(time.time() - start_time)

        # Time the O(2^n) function (exponential time)
        # For large n, this may take too long, so we only run it for smaller n (e.g., n <= 20)
        if n <= 20:
            start_time = time.time()
            exponential_time_example(n)
            results["Exponential"].append(time.time() - start_time)
        else:
            results["Exponential"].append(None)  # Mark as None for large n to avoid long runtimes

    return input_sizes, results

# Measure and plot the results
input_sizes, results = measure_execution_times()

import matplotlib.pyplot as plt

# Plotting the execution times
plt.figure(figsize=(12, 8))

# Plot each time complexity
plt.plot(input_sizes, results["Constant"], label="O(1) - Constant", marker='o', linestyle='--')
plt.plot(input_sizes, results["Logarithmic"], label="O(log n) - Logarithmic", marker='o')
plt.plot(input_sizes, results["Linear"], label="O(n) - Linear", marker='o')
plt.plot(input_sizes, results["N Log N"], label="O(n log n) - N Log N", marker='o')
plt.plot(input_sizes, results["Quadratic"], label="O(nÂ²) - Quadratic", marker='o')

# Exponential complexity will be plotted only for smaller input sizes
plt.plot(input_sizes[:21], results["Exponential"][:21], label="O(2^n) - Exponential", marker='o', linestyle=':')

# Labels and title
plt.title('Execution Time for Different Time Complexities')
plt.xlabel('Input Size (n)')
plt.ylabel('Execution Time (seconds)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Print the results for each complexity type
for complexity in results:
    print(f"{complexity}:")
    for n, time_taken in zip(input_sizes, results[complexity]):
        print(f"  n = {n}: {time_taken:.6e} seconds")
```
